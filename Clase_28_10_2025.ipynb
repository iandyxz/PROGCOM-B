```python
class Casa:
  def __init__(self, color, num_habitaciones, num_baos, tiene_patio, tiene_parqueadero):
    """
    Constructor de la clase casa
    Argumentos:
      color(str): definir el color de la casa
      num_habitaciones(int): nmero de habitaciones
      num_baos(int): nmero de baos
      tiene_patio(bool): si la casa posee patio
      tiene_parqueadero(bool): si la casa tiene parqueadero
    """
    self.color=color
    self.num_habitaciones=num_habitaciones
    self.num_baos=num_baos
    self.tiene_patio=tiene_patio
    self.tiene_parqueadero=tiene_parqueadero

  def describir(self):
    """
    Mtodo para describir la casa
    """
    descripcin=f"La casa es de color {self.color}. Tiene {self.num_habitaciones} habitaciones, y {self.num_baos} baos."

    if self.tiene_patio:
      descripcin+="\nLa casa posee un patio."
    if self.tiene_parqueadero:
      descripcin+="\nLa casa posee un parqueadero."
    return descripcin
  def pintar(self, nuevo_color):
    """
    Mtodo para pintar la casa
    Argumentos:
      nuevo_color(str): nuevo color de la casa
    """
    self.color=nuevo_colorprint(f"La casa ha sido pintada de color {self.color}.")
```

```python
class Floristera: #Herencia
  def __init__(self, color, num_habitaciones, num_baos, tiene_patio, tiene_parqueadero, nombre, flores):
    """
    Constructor de la clase floristera
    Argumentos:
      color(str): definir el color de la floristera
      num_habitaciones(int): nmero de habitaciones
      num_baos(int): nmero de baos
      tiene_patio(bool): si la floristera posee patio
      tiene_parqueadero(bool): si la floristera tiene parqueadero
      nombre(str): el nombre de la floristera
      flores(list): lista de flores
    """
    super().__init__(color, num_habitaciones, num_baos, tiene_patio, tiene_parqueadero)
    self.nombre=nombre
    self.flores=flores

  def describir_floristera(self):
    descripcin_casa=super().describir()
    descripcin_floristera=f"\nLa floristera se llama {self.nombre} y vende las siguientes flores: {', '.joing(self.flores)}."
    return descripcin_casa+descripcin_floristera
```

```python
mi_floristera=Floristera("amarilla", 1, 1, False, False, "Floristera", ["tulipn", "rosas", "girasoles"])
```
